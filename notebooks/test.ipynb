```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns only
numeric_df = df.drop(columns=['A2BBX6', 'A', 'B1', 'B2', 'X'])

# Compute correlation matrix
corr = numeric_df.corr()

# Increase figure size and rotate labels
plt.figure(figsize=(18, 14))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, square=False,
            annot_kws={"size":7})
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.title("Correlation Heatmap of Bandgap & Formation Energy Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
X_train_bandgap = df_bandgap.drop(columns=['band_gap'])
Y_train_bandgap = df_bandgap['band_gap']
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor

# Base model: a decision tree with max depth 10
base_model = DecisionTreeRegressor(max_depth=10)

# Bagging ensemble: 50 decision trees
model = BaggingRegressor(
    base_estimator=base_model,
    n_estimators=50,
    random_state=42
)

# Fit the model
model.fit(X_train_bandgap, Y_train_bandgap)
```
